```python
marks=int(input("Enter your marks: "));
if marks>=97:
    print("Grade A+");
elif marks>=93 and marks<97:
    print("Grade A");
elif marks>=90 and marks<93:
    print("Grade A-");
elif marks>=87 and marks<90:
    print("Grade B+");
elif marks>=83 and marks<86:
    print("Grade B");
elif marks>=80 and marks<83:
    print("Grade B-");
elif marks>=77 and marks<80:
    print("Grade C+");
elif marks>=73 and marks<77:
    print("Grade C");
elif marks>=70 and marks<73:
    print("Grade C-");
elif marks>=67 and marks<70:
    print("Grade D+");
elif marks>=63 and marks<67:
    print("Grade D");
elif marks>=60 and marks<63:
    print("Grade E-");
else:
    print("Grade F");   
```
